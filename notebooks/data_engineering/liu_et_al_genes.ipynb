```python
import pandas as pd
import os
import glob
import sys
import gget
import mygene
```

```python
fpath = "/nfs/turbo/umms-indikar/shared/projects/cell_cycle/data/liu_2017/GSE94479_GROCountNorm.txt.gz"

df = pd.read_csv(fpath, sep=" ")
gene_map = dict(zip(df['refseq'].values, df['id'].values))
df.head()
```

```python
dpath = "/nfs/turbo/umms-indikar/shared/projects/cell_cycle/data/liu_2017/"

file_list = glob.glob(f"{dpath}*Deseq*")

padj = 0.05
log2FoldChange = 1.0

result = []

for fpath in file_list:
    basename = os.path.basename(fpath).replace(".txt.gz", "")
    phase = basename.split("_")[-1]

    df = pd.read_csv(fpath, sep='\t')
    df = df.drop(columns=['Unnamed: 0', 'pvalue', 'stat'])
    
    df = df[df['padj'] < padj]
    df = df[df['log2FoldChange'] > log2FoldChange]
    df['phase'] = phase
    df = df.sort_values(by='log2FoldChange', ascending=False)
    
    df = df.reset_index(drop=True)
    df['gene_name'] = df['id'].map(gene_map)
    print(f"{df.shape=}")
    
    result.append(df)
    
    
result = pd.concat(result)
outpath = "../../resources/liu_2017_genes.csv"
result.to_csv(outpath, index=False)
result.head()
```

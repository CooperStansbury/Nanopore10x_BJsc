```python
import glob
import os
import sys
import subprocess
```

```python
# rpath = "/nfs/turbo/umms-indikar/shared/projects/cell_cycle/data/sc_cell_cycle_2024/transcripts/2024-08-19-GEX-Runs-2-to-5a/"
rpath = "/nfs/turbo/umms-indikar/shared/projects/cell_cycle/data/sc_cell_cycle_2024/transcripts/2024-08-23-GEX-4b-c-d-5b-c-6-7/"
outpath = "/nfs/turbo/umms-indikar/shared/projects/cell_cycle/data/sc_cell_cycle_2024/transcripts/merged/"

print("**** Starting to merge FASTQ files ****")

for d in os.listdir(rpath):

    dpath = f"{rpath}{d}/fastq_pass"
    print(f"\nProcessing directory: {d}")
    print(f"  FASTQ files path: {dpath}")
   
    files = glob.glob(f"{dpath}/*.fastq.gz")
    file_str = " ".join(files)

    outfile = f"{outpath}{d}.fastq.gz"
   
    print(f"  Merging into: {outfile}")
   
    # Build the command list
    command = [
        'cat',
        file_str,
        ">",
        outfile,
    ]
   
    # Execute the command using subprocess.run
    output = subprocess.run(" ".join(command), shell=True, capture_output=True)

    if output.returncode == 0:
        print(f"  Merge successful!")
    else:
        print(f"  Error during merge. Check the following output:")
        print(output.stderr.decode())

    # break  # Remove this line if you want to process all directories

print("**** Merging complete ****")
```
